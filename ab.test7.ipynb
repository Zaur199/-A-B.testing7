# РЕШЕНИЕ

В первом вебинаре в качестве описания был взят продукт ozon.ru. Все дальнейшие исследования будем делать на основе этого продукта.

Сформулируем А/В тесты.

***Пример A/B-теста №1:***

*Гипотеза: изменение цвета кнопки "Купить" на карточке товара увеличит конверсию покупок.*

***План эксперимента:***
1. Создание двух групп (контрольная и экспериментальная).
2. В контрольной группе цвет кнопки "Купить" останется прежним (например, зеленый), а в экспериментальной группе цвет будет изменен (например, на оранжевый).
3. Случайное распределение пользователей на две группы.
4. Сбор данных в течение определенного периода времени (например, 2 недели).
5. Анализ данных с помощью статистических методов (например, t-тест) для выявления статистически значимой разницы между группами.

***Метрики для анализа:***
- Конверсия покупок (доля пользователей, совершивших покупку относительно общего количества посетителей);
- Количество покупок в каждой группе;
- Средний чек.

***Пример A/B-теста №2:***

*Гипотеза: изменение порядка фильтров на странице категории товаров с учетом их популярности улучшит пользовательский опыт и время на сайте.*

***План эксперимента:***
1. Создание двух групп (контрольная и экспериментальная).
2. В контрольной группе порядок фильтров останется прежним, а в экспериментальной порядок будет изменен в соответствии с популярностью.
3. Случайное распределение пользователей на две группы.
4. Сбор данных в течение определенного периода времени (например, 1 месяц).
5. Анализ данных с помощью статистических методов (например, t-тест) для выявления статистически значимой разницы между группами.

***Метрики для анализа:***
- Среднее время пребывания на странице категории товаров;
- Количество товаров, добавленных в корзину;
- Количество совершенных покупок;
- Средний чек.
